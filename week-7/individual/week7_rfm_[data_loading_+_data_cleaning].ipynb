```python
import pandas as pd

df_sales = pd.read_csv("datasets/clean/sales.csv")
df_customers = pd.read_csv("datasets/clean/customers.csv")

print("SALES")
print(df_sales.shape)
print(df_sales.dtypes)
print(df_sales.head())

print("\nCUSTOMERS")
print(df_customers.shape)
print(df_customers.dtypes)
print(df_customers.head())

df = pd.merge(df_sales, df_customers, on="customer_id", how="left")

print("MERGED")
print(df.shape)
print(df.dtypes)
print(df.head())
print(df.columns)
```

```python
print("INITIAL SHAPE:", df.shape)

print("Duplicates:", df.duplicated().sum())
df = df.drop_duplicates()

print("\nNULL values per column:")
print(df.isnull().sum())

df = df.dropna(subset=["customer_id", "sale_date", "total_price"])

df["sale_date"] = pd.to_datetime(df["sale_date"], format="mixed", dayfirst=True, errors="coerce")

df = df.dropna(subset=["sale_date"])

df = df[df["total_price"] > 0]

print("\nCLEANING REPORT")
print("Final shape:", df.shape)
print("Unique customers:", df["customer_id"].nunique())
print("Date range:", df["sale_date"].min(), "to", df["sale_date"].max())

df.head()
```
